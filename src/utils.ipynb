```python
import pandas as pd
input_path_descr = '/Users/thebekhruz/Desktop/100Days-Of-Code/100-Days-of-NLP-Odyssey/data/intermediate/preprocessing/un_lemmatised/processed_data_descr_no_lemma.csv'
input_path_tile = '/Users/thebekhruz/Desktop/100Days-Of-Code/100-Days-of-NLP-Odyssey/data/intermediate/preprocessing/un_lemmatised/processed_data_title_no_lemma.csv'
```

```python
df_descr = pd.read_csv(input_path_descr, delimiter='\t', header=None, names=['doc_id', 'type', 'value'])
df_title = pd.read_csv(input_path_tile, delimiter='\t', header=None, names=['doc_id', 'type', 'value'])

df_title.head()
```

```python
combined_df = pd.merge(df_title, df_descr, on='doc_id', suffixes=('_title', '_description'))
# combined_df.head()

# Concatenate the title and description values
combined_df['combined_value'] = combined_df['value_title'] +  ". " + combined_df['value_description']
combined_df.head()
```

```python
columns_to_drop = ['type_title', 'value_title', 'type_description', 'value_description']
combined_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
output_file_path = '/Users/thebekhruz/Desktop/100Days-Of-Code/100-Days-of-NLP-Odyssey/data/intermediate/preprocessing/combined_df_for_lda.csv'
combined_df.to_csv(output_file_path, sep='\t', index=False, header=None)
```
